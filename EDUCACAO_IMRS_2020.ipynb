# Dados relacionados as **taxas de analfabetismo**

---






### **Filtragem de mesorregiões**

```python
# Fitro que puxa dados das mesorregiões "Norte de Minas" e "Sul/Sudoeste de Minas" para serem comparadas
df_filtro = df[(df.nome_meso == "Norte de Minas") | (df.nome_meso == "Sul/Sudoeste de Minas")]
df_filtro.head()
```

```python
# Filtrando apenas a região do Norte de Minas
df_norte_minas = df[(df.nome_meso == "Norte de Minas")]
```

```python
# Filtrando apenas a região do Sul/Sudoeste de Minas
df_sul_minas = df[(df.nome_meso == "Sul/Sudoeste de Minas")]
```

###**Medidas-Resumo**

```python
# Analisando as principais medidas-resumo da taxa de analfabetismo
df.tx_analf_25m.describe()
```

**Medidas de posição ou centralidade**

```python
# Pegando a moda da taxa de analfabetismo e arredondando para duas casas
df.tx_analf_25m.mode().round(2)
```

```python
# Pegando a média da taxa de analfabetismo e arredondando para duas casas
df.tx_analf_25m.mean().round(2)
```

**Medidas de dispersão ou de variabilidade**

```python
# Pegando o desvio padrão da taxa de analfabetismo e arredondando para duas casas
round(df.tx_analf_25m.std(), 2)
```

**Medidas de separação ou quantis**

```python
# Pegando o Xmax da taxa de analfabetismo no Sul de Minas e arredondando para duas casas
df_sul_minas.tx_analf_25m.quantile(1).round(0)
```

```python
# Pegando o primeiro quartil da taxa de analfabetismo no Norte de Minas e arredondando para duas casas
df_norte_minas.tx_analf_25m.quantile(0.25).round(0)
```

### **Distribuição de frequências**

```python
# Distribuição da taxa de analfabetismo da população mineira acima dos 25 anos

plt.figure(figsize=(10, 6)) # Definindo o tamanho do histograma
sns.histplot(data=df, x=df.tx_analf_25m, color="red", binwidth=2.58) # Gerando histograma a partir do eixo x, de cor vermelha e barras com largura definida

# Definindo o título do histograma
plt.title("Distribuição da taxa de analfabetismo da população acima dos 25 anos em MG", fontsize=13)

# Definindo títulos dos eixos x e y
plt.xlabel("Taxa de analfabetismo (%)")
plt.ylabel("Quantidade de municípios")

plt.xticks([5, 10, 15, 20, 25, 30, 35, 40, 45]) # Definindo os elementos do eixo x

plt.show() # Plotando o gráfico de forma limpa
```

### **Boxplot ou diagrama de caixa-de-bigode**

**Comparação de todas as regiões mineiras**

```python
# Boxplot que traz a distribuição da taxa de analfabetismo por cada mesorregião

plt.figure(figsize=(14, 6)) # Definindo o tamanho do diagrama

# Gerando boxplot a partir da definição dos eixos x e y, seperando cada caixa com uma cor especifica, removendo todos os outliers da visualização
sns.boxplot(x=df.nome_meso, y=df.tx_analf_25m, hue=df["nome_meso"], fliersize=0)

# Definindo o título do diagrama
plt.title("Distribuição da Taxa de Analfabetismo por Mesorregião", fontsize=15)

# Definindo títulos dos eixos x e y
plt.xlabel("Mesorregião")
plt.ylabel("Taxa de Analfabetismo (%)")

plt.yticks([5, 10, 15, 20, 25, 30, 35, 40, 45]) # Definindo os elementos do eixo y

plt.xticks(fontsize=8, rotation=30) # Definindo a formatação do eixo x, como fonte e rotação

plt.show() # Plotando o gráfico de forma limpa
```

**Comparação com as regiões filtradas**

```python
# Boxplot que traz a distribuição da taxa de analfabetismo de regiões previamente filtradas

plt.figure(figsize=(14, 6)) # Definindo o tamanho do diagrama

# Gerando boxplot a partir da definição dos eixos x e y, removendo todos os outliers da visualização
sns.boxplot(x=df_filtro.nome_meso, y=df_filtro.tx_analf_25m, fliersize=0)

# Definindo o título do diagrama
plt.title("Taxa de Analfabetismo nas Mesorregiões Filtradas", fontsize=15)

# Definindo títulos dos eixos x e y
plt.xlabel("Mesorregião", fontsize=12)
plt.ylabel("Taxa de Analfabetismo (%)", fontsize=12)

plt.xticks(fontsize=11) # Definindo tamanho do titulo dos elementos do eixo x

plt.show() # Plotando o gráfico de forma limpa
```
